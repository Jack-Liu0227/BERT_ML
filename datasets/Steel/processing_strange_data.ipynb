```python
# 读取数据
import pandas as pd
import numpy as np
import os
csv_path="data.csv"
data=pd.read_csv(csv_path)
display(data.info())
```

# 列名重命名

```python
col_dict={}
for ele in data.columns[12:48]:
    col_dict[ele]=ele+"(wt%)"

col_dict.update({
    "Tensile_value":"UTS(MPa)",
    "Yield_value":"YS(MPa)",
    "Elongation_value":"El(%)",
    "Text":"Processing_Description"
})

data.rename(columns=col_dict,inplace=True)
data.info()
```

```python
data.to_csv("USTB_steel_processed.csv",index=False)
```
