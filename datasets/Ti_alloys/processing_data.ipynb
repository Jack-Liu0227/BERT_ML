# 读取TNCH_dataset_0530.xlsx数据

```python
# 读取Excel文件
import pandas as pd

# Read the Excel file
df = pd.read_excel('TNCH_dataset_0530.xlsx')

# Print the first two columns
print("First two columns of the dataset:")
display(df.head())
display(df.info())
```

# 数据清洗

```python
# 将第二行的单位信息合并到第一行的列名中
# Merge units from second row into first row column names
new_columns = []
for col1, col2 in zip(df.iloc[0], df.iloc[1]):
    if pd.isna(col2):
        new_columns.append(str(col1))
    
    elif col2 == '%' and col1!="cold-roling-amount":
        new_columns.append(f"{col1}(wt%)")
    else:
        new_columns.append(f"{col1}({col2})")

# 设置新的列名并删除第二行
# Set new column names and remove the second row
df.columns = new_columns
display(df.head(2))
df = df.drop(index=[0, 1]).reset_index(drop=True)

# 打印处理后的前两行数据
# Print first two rows after processing
print("\nProcessed data first two rows:")
display(df.head())
display(df.info())
```

# 保存数据

```python
df.to_csv("TNCH_dataset_0530_cleaned.csv",index=False)
```
