## 数据清理


```python
import pandas as pd
import numpy as np

# 读取CSV文件
# Read CSV file
csv_path = './processed_acta_materialia_data.csv'
df = pd.read_csv(csv_path)

# 重命名前14列，将/％替换为wt%
# Rename first 14 columns, replace /％ with wt%
for col in df.columns[:10]:
    try:
        new_col = col.replace('/％', '(wt%)')
        df.rename(columns={col: new_col}, inplace=True)
    except:
        print(f"Error renaming column: {col}")
for col in df.columns[11:14]:
    try:
        new_col = col.replace('/%', '(wt%)')
        df.rename(columns={col: new_col}, inplace=True)
    except:
        print(f"Error renaming column: {col}")
df.rename(columns={'RE/%': 'Re(wt%)'}, inplace=True)


# 显示数据框信息
# Display dataframe info
df.info()
```
